```python
import copy
wav, sr = torchaudio.load(audio_path)
resampled_wav = f.resample(wav, sr, 16000, rolloff=0.99)
filtered_wav = copy.deepcopy(resampled_wav)
for i in range(3):
    filtered_wav = f.bandpass_biquad(filtered_wav, 16000, 100, 100/200)
#filtered_wav = resampled_wav - filtered_wav
mel = torchaudio.transforms.MelSpectrogram(16000)(filtered_wav)
mel_original = torchaudio.transforms.MelSpectrogram(16000)(resampled_wav)
```

```python
def plot_spectrogram(specgram, title=None, ylabel="freq_bin", ax=None):
    if ax is None:
        _, ax = plt.subplots(1, 1)
    if title is not None:
        ax.set_title(title)
    ax.set_ylabel(ylabel)
    ax.imshow(librosa.power_to_db(specgram), origin="lower", aspect="auto", interpolation="nearest")
plot_spectrogram(mel[0], title="MelSpectrogram - torchaudio", ylabel="mel freq")
```

```python
plot_spectrogram(mel_original[0], title="MelSpectrogram - torchaudio", ylabel="mel freq")
```
